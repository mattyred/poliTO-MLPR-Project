# Data Inspection

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constant values
NFEATURES = 12 # Num Features
NTSAMPLES = 6000 # Num Training Samples
```

```python
# Dataset reading
DT = np.zeros(shape=(NFEATURES,NTSAMPLES), dtype='float32') # DT: Data Training
LT = np.zeros(NTSAMPLES, dtype='int32') # LT: Labels Training
with open('../Data/Train.txt','r') as file:
    i = 0
    for line in file:
        features_list = line.split(',')
        features = np.array(features_list[0:NFEATURES], dtype='float32').reshape(-1,1)
        label = int(features_list[-1])
        DT[:,i:i+1] = features
        LT[i] = label
        i += 1
```

```python
DT_male = DT[:,LT==0]
DT_female = DT[:,LT==1]
for i in range(NFEATURES):
    plt.figure()
    plt.hist(DT_male[i,:], bins=10, alpha=0.4, ec='black')
    plt.hist(DT_female[i,:], bins=10, alpha=0.4, ec='black')
    plt.legend(['male','female'])
```

```python
print('Minum value for each feature: %s'%([min(DT[i,:]) for i in range(NFEATURES)]))
```

```python
print('Maximum value for each feature: %s'%([max(DT[i,:]) for i in range(NFEATURES)]))
```

```python
print('Mean for each feature: %s'%([np.mean(DT[i,:]) for i in range(NFEATURES)]))
```

```python
print('Variance for each feature: %s'%([np.var(DT[i,:]) for i in range(NFEATURES)]))
```
